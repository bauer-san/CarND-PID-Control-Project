```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
csv = np.genfromtxt ('./build/output.txt', delimiter=",")
#cte = csv[:,0]
p_error = csv[:,1]
i_error = csv[:,2]
d_error = csv[:,3]
steer_value = csv[:,4]

Kp = 0.12
Ki = 0.001
Kd = 1.1

plt.figure(figsize=(12, 8))
#plt.plot(cte)
plt.plot(Kp*p_error)
plt.plot(Ki*i_error)
plt.plot(Kd*d_error)
plt.plot(steer_value)
plt.legend(['P', 'I','D', 'steer'])
plt.title('Cross-Track Error controller')
```

```python
SPEEDp_error = csv[:,6]
SPEEDi_error = csv[:,7]
SPEEDd_error = csv[:,8]
throttle_cmd = csv[:,9]

Kp=0.075
Ki=0.0001
Kd=0.

plt.figure(figsize=(12, 8))
plt.plot(Kp*SPEEDp_error)
plt.plot(Ki*SPEEDi_error)
plt.plot(Kd*SPEEDd_error)
plt.plot(throttle_cmd)
plt.legend(['P', 'I','D', 'throttle'])
plt.title('Speed Error controller')
```
